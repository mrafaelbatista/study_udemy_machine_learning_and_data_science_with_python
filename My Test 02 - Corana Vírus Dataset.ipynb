```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime
```

```python
covid_brasil = pd.read_excel('./HIST_PAINEL_COVIDBR_17jun2020.xlsx')
```

```python
covid_brasil.head()
```

```python
covid_jp = covid_brasil.loc[covid_brasil['municipio'] == 'João Pessoa']
```

```python
covid_jp.head()
```

```python
covid_jp.describe()
```

```python
# Preparando os labels do eixo x
one_day = datetime.timedelta(days=1)

datas = covid_jp['data']
obitos_acumulados = covid_jp['obitosAcumulado']

# Ainda não entendo bem o motivo da linha abaixo existir
fig, ax = plt.subplots()

plt.style.use('grayscale')
plt.plot(datas, obitos_acumulados, 'g-', linewidth=2)

plt.xticks(datas[::2], rotation=90)

fig.figure(figsize=(9, 3))

# Título do Gráfico e Nomes dos Eixos
plt.title('Gráfico 1 - Número de óbitos acumulados por dia', fontsize=16)
plt.ylabel('Óbitos Acumulados')
plt.xlabel('Dia')

plt.show()
```
