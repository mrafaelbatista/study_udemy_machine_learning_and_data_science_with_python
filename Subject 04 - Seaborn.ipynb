# Seaborn

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(1, 10, 100)
```

```python
plt.plot(x, np.sin(x), 'b--')
plt.plot(x, np.cos(x), 'r--')
plt.show()
```

```python
import seaborn as sns
sns.set()
```

```python
plt.plot(x, np.sin(x), 'b--')
plt.plot(x, np.cos(x), 'r--')
plt.show()
```

```python
sns.set()
sns.set_style('ticks')
plt.plot(x, np.sin(x), 'b--')
plt.plot(x, np.cos(x), 'r--')
plt.show()
```
